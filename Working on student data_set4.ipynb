```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
from sklearn.metrics import mean_squared_error
df = pd.read_csv(r"X:\sir ismail AI course\jupyter data sice\dataset\Student_Marks.csv")
df
```

```python
plt.plot(df["Marks"])
```

```python
plt.scatter(df["Marks"],df["number_courses"])
plt.xlabel("Marks")
plt.ylabel("cources")
plt.show()
```

```python
df[df["number_courses"] == 3]
```

```python
df[df["time_study"] == max(df["time_study"])]
```

```python
df[(df["Marks"] > 10) & (df["number_courses"] == 3)]
```

```python
X = df[["number_courses","Marks"]] # independent varibales
y = df["time_study"] # dependent varialbe
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=51, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train =  sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
X_train
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([X_test[2]])
```

```python
model.score(X_test,y_test) # r square
```

```python
y_predict = model.predict(X_test)
```

```python
# Mean Square Error
mse = mean_squared_error(y_test,y_predict)
# Root mean Square error
rmse = np.sqrt(mse)
```
